```python
N = 20
lattice = SquareLattice(N)
lattice.canonical_order()
lattice.remove_sites([N // 2 + i * N + j for i in range(N//2) for j in range(N//2)])
new_bonds = lattice.glue_bond([i * N // 2 + N // 2 - 1 for i in range(N//2)], [(N//2)**2 + N - 1 - i for i in range(N//2)], [2, 0], codim=1)

print("new bonds = ", new_bonds)
lattice.plot(True)
```
